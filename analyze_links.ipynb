```python
import matplotlib.pyplot as plt
import pandas as pd
import json
```

```python
with open("link_domains2.json") as ff:
    domains = json.load(ff)
```

```python
domains_flat = [item for sublist in domains for item in sublist]
df = pd.DataFrame(domains_flat, columns=["link"])
```

```python
df.link.value_counts()
```

```python
df.link.value_counts()[:100].plot(kind='bar', figsize=(20, 10))
```

```python
remove_list = ["bit.ly", "t.me", "amazon", "afkstore", "dft", "indieweb", "spotify", "youtu", "shooter", "on.fb", "twitch", "heya", "twitter", "breakingitaly","amzn", "youtube", "patreon", "facebook", "instagram"]
pattern = '|'.join(remove_list)
ddf = df[~df['link'].str.contains(pattern)]
```

```python
ddf.link.value_counts()
```

```python
ddf.link.value_counts()[:30].plot(kind='bar')
```
